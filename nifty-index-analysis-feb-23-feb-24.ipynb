# Loading Required Libraries and Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/kaggle/input/various-nifty-indices-tsa-data-14-02-2023-24/Nifty50.csv")
data
```

```python
# type of date column
type(data.date[0])
```

```python
# Converting the type from str to TimeStamp
data.date = pd.to_datetime(data.date)
type(data.date[0])
```

```python
#Setting date as index
data.set_index('date',inplace=True)
```

```python
# Removing the index name
data.index.name = None
data
```

## Overall View

```python
data['2023':'2024'].plot(subplots=True, figsize=(10,12))
```

```python
sns.lineplot(data , x = data.index , y = "open" , color = 'b')
plt.title('Periods By Open Index')
plt.xlabel('Period')
plt.ylabel('Opening Index')
plt.grid(True)
plt.show()
```

```python
sns.lineplot(data , x = data.index , y = "close" , color = 'g')
plt.title('Periods By Close Index')
plt.xlabel('Period')
plt.ylabel('Closing Index')
plt.grid(True)
plt.show()
```

```python
#data.turnover.plot(figsize=(10,5))
sns.lineplot(data , x = data.index , y = "turnover" , color = 'r')
plt.title('Periods By Turnover')
plt.xlabel('Period')
plt.ylabel('Turnover')
plt.grid(True)
plt.show()
```

## Opening Prices and Close Price of fistmonth

```python
# head 30

Opening_Price_1st_Month = data.head(30)
plt.figure(figsize=(10, 6))
sns.lineplot(x=Opening_Price_1st_Month.index,y='open',data=Opening_Price_1st_Month,label="open", color = 'r' , marker='o')
sns.lineplot(x=Opening_Price_1st_Month.index,y='close',data=Opening_Price_1st_Month,label="close", color = 'y' , marker='o')
plt.legend()
```

## Opening Prices and Close Price of Lastmonth

```python
# tail 30

Opening_Price_Last_Month = data.tail(30)
plt.figure(figsize=(10, 6))
sns.lineplot(x=Opening_Price_Last_Month.index,y='open',data=Opening_Price_Last_Month,label="open", color = 'g' , marker='o')
sns.lineplot(x=Opening_Price_Last_Month.index,y='close',data=Opening_Price_Last_Month,label="close", color = 'b' , marker='o')
plt.legend()
```

## Monthly High

```python
Mnth_High = data.resample('M').max()
plt.figure(figsize=(10, 6))
plt.xticks(rotation = 45)
sns.barplot(data=Mnth_High,x=Mnth_High.index,y="high")
plt.title('Monthly High')
plt.xlabel('Period')
plt.ylabel('High')
plt.show()
```

## Monthly Low

```python
Mnth_Low = data.resample('M').min()
plt.figure(figsize=(10, 6))
plt.xticks(rotation = 90)
sns.barplot(data=Mnth_Low,x=Mnth_Low.index,y="low")
plt.title('Monthly Low')
plt.xlabel('Period')
plt.ylabel('Low')
plt.show()
```

## Number of shares Traded

```python
No_ST = data.resample('M').sum()
plt.figure(figsize=(10, 6))
plt.xticks(rotation = 90)
sns.barplot(data=No_ST,x=No_ST.index,y="shares traded")
plt.title('Number of Shares Traded')
plt.xlabel('Period')
plt.ylabel('Shares Traded')
plt.show()
```
